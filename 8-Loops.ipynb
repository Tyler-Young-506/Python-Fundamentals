```python
canadianPrimeMinisters = [
    "Justin Trudeau",
    "Stephen Harper",
    "Paul Martin",
    "Jean Chretien",
    "Kim Campbell",
    "Brian Mulroney"
]

for primeMinister in canadianPrimeMinisters:
    print(primeMinister)
print("")
for x in canadianPrimeMinisters:
    print(x)
print("")
for smellyCat in canadianPrimeMinisters:
    print(smellyCat)
print("")
for cliffordTheDog in canadianPrimeMinisters:
    print(cliffordTheDog)
print("")
```

```python
#While loops will run as long as some external condition is true.
#Infinite loops occur when the external condition is always true. Example:

bool = True

while(bool):
    print("This is an infinite loop, and this will print A LOT.")
```

```python
#Variables (usuall i, j, and k) are defined outside of a while loop and an increment operator will be used within the while loop to increase the value of that variable.

i = 1
i += 1
print(i)
```

```python
#Example of a while loop:

numberString = ""

i = 0

while(i<10):
    numberString = numberString + str(i)
    i += 1
print(numberString)
```

```python
#Sometimes, problems can be solved using either a for loop or a while loop. Example:

ingredientList = ['mushrooms', 'bacon', 'green peppers', 'pepperoni', 'cheese']

#For loop
for ingredient in ingredientList:
    print(ingredient)
    
print("")   

#While loop
i = 0
while i < len(ingredientList):
    print(ingredientList[i])
    i += 1
```

```python
# Role Call Example:

students = ['Nick', 'John', 'Joe']

#For loop
for student in students:
    print(f"Is {student} here?")

print("")

#While loop
i = 0
while i < len(students):
    print(f"Is {students[i]} here?")
    i += 1
```
